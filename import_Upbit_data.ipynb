# Import Upbit BTC data using its API

```python
import pandas as pd
from datetime import datetime
```

```python
import requests

url = "https://api.upbit.com/v1/candles/days"
days = 10000
count = 200
last_date = '2019-06-24' # Everyday at 09:00 KST

for i in range(days/count):
    querystring = {"market":"KRW-BTC", 'count':count, 'to':}

response = requests.request("GET", url, params=querystring)

pd.DataFrame(response.json())
```
