```python
df = pd.DataFrame()

url = "https://api.upbit.com/v1/candles/days"

days = 10000
count = 200
last_date = datetime.datetime.strptime('2019-06-24 09:00:00', '%Y-%m-%d %H:%M:%S') # Everyday at 09:00 KST

for i in range(days//count):
    print("processing {i}/{total}".format(i=i, total=days//count))
    td = datetime.timedelta(days=count)
    to = last_date - i*td
    querystring = {"market":"KRW-BTC", 'count':count, 'to':str(to)}

    response = requests.request("GET", url, params=querystring)
    
    
    df = df.append(pd.DataFrame(response.json()), ignore_index=True)
    
    if i%3 == 0 :
        time.sleep(1)

df
```
