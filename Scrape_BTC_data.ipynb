# Scraping BTC data from coinmarketcap.com

```python
import pandas as pd
import html5lib
```

```python
start = '20140101'
end = '20190613'

url = "https://coinmarketcap.com/ko/currencies/bitcoin/historical-data/?start={start}&end={end}".format(start=start, end=end)
```

```python
df = pd.read_html(url)[0]
df.head()
```

```python
df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume.USDT', 'MarketCap.USD']
df.head()
```

```python
df.to_csv('BTC_coinmarketcap.csv')
```
