# 한국투자증권 리서치 챌린지에서 받은 Excel파일들을 Pandas Dataframe으로 변환하고 정리한다. 

```python
import pandas as pd
```

Excel support를 위해 pandas 라이브러리가 내부적으로 사용하는 xlrd, openpyxl 설치가 필요하며 자료는 국내 자료만 쓰겠다. 

모든 시트를 따로따로 읽어들여 변수에 저장하고 컬럼 명을 영어로 바꾼다. 

```python
xlsx = pd.ExcelFile('./data_provided/korea.xlsx')
```

```python
macro_kor = pd.read_excel(xlsx, '매크로 지표_한국', skiprows=2)
macro_kor = macro_kor.drop(columns=['Unnamed: 0'])
macro_kor.rename({macro_kor.columns[0]:'date'}, axis=1, inplace=True)
macro_kor.columns = ['date', 
                     'real_GDP_YoY', 
                     'real_GDP_QoQ', 
                     'export_USDmn', 
                     'import_USDmn', 
                     'CPI_YoY', 
                     'BOK_manufacture_idx', 
                     'KOR_bond_3y', 
                     'KOR_bond_10y', 
                     'exchange_krw_usd',
                    ]
macro_kor.head()
```

```python
KOSPI_stats = pd.read_excel(xlsx, '코스피 지수 추이', skiprows=13)
KOSPI_stats.columns = ["date",
                       "close",
                       "revenue",
                       "operating_profits",
                       "net_profits",
                       "PER",
                       "PBR",
                       "EPS",
                      ]
KOSPI_stats.head()
```
